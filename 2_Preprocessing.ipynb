The goal of this notebook is to import the raw dataset and apply the transformations found during the 1_EDA.ipynb notebook to generate clean dataset(s) to train a set of models.

```python
import pandas as pd
```

```python
# Load dataset
df_cali = pd.read_csv("data/raw/california_housing.csv")

# Explore dataset
df_cali.head()
```

```python
# Drop Population and Average Occupancy
df_processed_v1 = df_cali.drop(columns=["Population", "AveOccup"])

# Check for missing values
df_processed_v1.head()
```

```python
# Drop Latitude and Longitude
df_processed_v2 = df_processed_v1.drop(columns=["Latitude", "Longitude"])

# Check for missing values
df_processed_v2.head()
```

```python
df_processed_v1.to_csv("data/processed/california_housing_processed_v1.csv", index=False)
df_processed_v2.to_csv("data/processed/california_housing_processed_v2.csv", index=False)
```
